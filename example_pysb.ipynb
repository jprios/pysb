The data for this example comes from the Kaggle database ["Customer purchase behavior - Electronic Sales Data"](kaggle.com/datasets/cameronseamons/electronic-sales-sep2023-sep2024/data)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm
import matplotlib.pyplot as plt
from pysb import pysb
```

```python
df = pd.read_csv(r"Electronic_sales_Sep2023-Sep2024.csv")
df['Purchase Date'] = pd.to_datetime(df['Purchase Date'])
#Create a daily time series of the Total Purchase from september 2023 to september 2024
series = df.groupby('Purchase Date')['Total Price'].sum()

#Plot the time series

plt.figure(figsize=(10, 6))
plt.plot(series, label='Time Series', color='blue')
plt.title('Time Series with Supremum Wald Test Breakpoint')
plt.xlabel('Time Index')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
sup_wald_stat, breakpoint, p_value = pysb(series)

print(f"Supremum Wald Statistic: {sup_wald_stat}")
print(f"Breakpoint date: {breakpoint}")
print(f"P-value: {p_value}")
```
